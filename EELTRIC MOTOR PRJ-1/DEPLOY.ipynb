```python
# Importing the libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the dataset (csv file)
df= pd.read_csv("/content/temperature_data (1).csv")
df.head()
```

```python
df
```

```python
# Dropping profile id as it does not have any significance in predicting motor speed
df_1=df.drop(['profile_id'],axis=1)
```

```python
# Splitting the dataset into dependent and independent variable
# Here we put every  variables except motor speed into X variablle and motor speed into y variable
X=df_1.drop(['motor_speed'],axis=1)
y=df_1['motor_speed']
y
```

```python
X
```

```python
#Splitting Data into train and test Data
from sklearn.model_selection import train_test_split,GridSearchCV
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
# Here we have used 70% data for training our model and kept 30% data for testing purpose
```

```python
from sklearn.tree import DecisionTreeRegressor
model_1=DecisionTreeRegressor()
model_1.fit(X_train,y_train)
```

```python
y_pred1=model_1.predict(X_test)
# Predection of y values on X test dataset
```

```python
# Evaluationg The Model with the help of r2 score matrix
from sklearn.metrics import r2_score
rms_1=np.round(np.mean((y_pred1-y_test)**2),5)
rmse_1=np.round(np.mean((y_pred1-y_test)**0.5),5)
rmsle_1=np.round(np.log(rmse_1),5)
r2_1=np.round(r2_score(y_test,y_pred1),5)
ar_1=np.round(1 - ((1-r2_1)*(299421-1)/(299421-11-1)),5)
print(rms_1,rmse_1,rmsle_1,r2_1,ar_1)
```

```python
model_1.score(X_test,y_test)
# Accuracy of the Model
```

```python
%%writefile app.py
import pickle
import streamlit as st

# open model in read binary mode /content/model_1.pkl
load = open('model_1.pkl','rb')
model = pickle.load(load)
def main():
    st.title('Motor Speed Prediction Model')
    # Accept input values from user through browser
    ambient = st.number_input('Ambient temperature as measured by a thermal sensor located closely to the stator: ')
    coolant = st.number_input('Coolant temperature: ')
    u_d= st.number_input('Voltage d-component: ')
    u_q = st.number_input('Voltage q-component: ')
    torque = st.number_input('Torque induced by current: ')
    i_d= st.number_input('Current d-component: ')
    i_q = st.number_input('Current q-component: ')
    pm = st.number_input('Permanent Magnet surface temperature: ')
    stator_yoke=st.number_input('Stator yoke temp:')
    stator_tooth=st.number_input('Stator tooth temp:')
    stator_winding=st.number_input('Stator winding temp:')
# Define predict function
    if st.button('Predict'):
        makepredictionprediction = model.predict([[ambient,coolant,u_d,u_q,torque,i_d,i_q,pm,stator_yoke,stator_tooth,stator_winding]])
        output=round(makepredictionprediction[0],2)
        st.success('The motor speed is {}'.format(output))
if __name__ == '__main__': # it will always be true so calls the main() function
        main()
```
